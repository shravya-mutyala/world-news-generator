```python
import pandas as pd
```

```python
df = pd.read_json('top_news_response.json', orient='records')
df
```

```python
# Step 1: Extract the list under the 'news' key in each 'top_news'
df['top_news'] = df['top_news'].apply(lambda x: x['news'])

# Step 2: Explode into individual news rows
df_exploded = df.explode('top_news').reset_index(drop=True)

# Step 3: Flatten the dictionaries inside each news item
news_flat = pd.json_normalize(df_exploded['top_news'])

# Step 4: Join back with the original metadata (language, country)
final_df = pd.concat([df_exploded.drop(columns=['top_news']), news_flat], axis=1)
```
